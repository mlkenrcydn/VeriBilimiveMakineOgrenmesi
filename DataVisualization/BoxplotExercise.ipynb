```python
df = sns.load_dataset("titanic")
```

```python
df.groupby('class')
```

```python
df_grouped = df.groupby('class')
```

```python
df_grouped.count()
```

```python
df_grouped.describe()
```

```python
df_grouped["fare"].mean()
```

```python
df_grouped["age"].max()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data= df)
plt.title("Age by class")
plt.xlabel("class")
plt.ylabel("age")

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data= df)
plt.title("Fare by class")
plt.xlabel("class")
plt.ylabel("fare")

plt.tight_layout() # grafikler sıkışık durduğunda düzeltmek için
plt.show()
```
