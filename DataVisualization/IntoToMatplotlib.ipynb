```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,40,40,50,60,70,80]
```

```python
weight_list = [25,55,58,60,66,56,55,70,70,75]
```

```python
plt.plot(age_list, weight_list, "b") # y-> yellow, g, r, b
# plt.savefig("test.png")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list, "y")
plt.title("Age-Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20). # 0 dan 10 a kadar 20 sayıyı eşit aralıkta listele
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "g*-")
plt.show() # üssel artan grafik
```

```python
plt.subplot(1,2,1) # (row, column, 1. grafik)
plt.plot(numpy_arr1, numpy_arr2, "b*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "y--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.7,0.7])
figure_axes.plot(numpy_arr1, numpy_arr2)
figure_axes.set_xlabel("x Axis")
figure_axes.set_ylabel("y Axis")
plt.show()
```

```python
new_fig = plt.figure(dpi=120) # dpi çözünürlük

new_axes = new_fig.add_axes([0.1,0.1,0.6,0.6])

new_axes.plot(numpy_arr1, numpy_arr1**2, label="numpy array **2")
new_axes.plot(numpy_arr1, numpy_arr1**3, label="numpy array **3")
new_axes.legend() # loc=1 -> labelin konumunu belirler
plt.show()
```
