```python
plt.scatter("Height", "Weight", data= df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
# Seaborn themes - search
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", data= df)
plt.xlabel("Height of Athletic")
plt.ylabel("Weight of Athletic")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", data= df) # hue vurgulanmak istenen colonu kullanÄ±r
plt.xlabel("Height of Athletic")
plt.ylabel("Weight of Athletic")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", data= df) # hue vurgulanmak istenen colonu kullanÄ±r
plt.xlabel("Height of Athletic")
plt.ylabel("Weight of Athletic")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", size="Age", data= df) # hue vurgulanmak istenen colonu kullanÄ±r
plt.xlabel("Height of Athletic")
plt.ylabel("Weight of Athletic")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lineplot(x="Height", y="Weight", hue="Sex", data= df)
plt.xlabel("Height of Athletic")
plt.ylabel("Weight of Athletic")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height", hue="Sex", data= df)
plt.ylabel("Frequency") # gÃ¶rÃ¼lme sÄ±klÄ±ÄŸÄ±
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height", hue="Sex", data= df, kind="kde")
plt.ylabel("Frequency") # gÃ¶rÃ¼lme sÄ±klÄ±ÄŸÄ±
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal", y="Height", hue="Sex", data=df)
plt.title("Medal of Height")
plt.show()
```
