```python
df1m = pd.read_csv('7-merge_data1.csv')
```

```python
df2m = pd.read_csv('7-merge_data2.csv')
```

```python
df1m
```

```python
df2m
```

```python
# merge - outer join
```

```python
df_merged_outher = pd.merge(df1m, df2m, on="Employee_ID", how="outer") # outher tüm datayı saklayıp eid ye göre eşleşenleri yazdırır
```

```python
df_merged_outher
```

```python
# merge - left join
```

```python
df_merge_left = pd.merge(df1m, df2m, on="Employee_ID", how="left")
```

```python
df_merge_left
```

```python
# merge - right join
```

```python
df_merge_right = pd.merge(df1m, df2m, on="Employee_ID", how="right")
```

```python
df_merge_right
```

```python
# merge - inner join
```

```python
df_merge_inner = pd.merge(df1m, df2m, on="Employee_ID", how="inner")
```

```python
df_merge_inner
```
