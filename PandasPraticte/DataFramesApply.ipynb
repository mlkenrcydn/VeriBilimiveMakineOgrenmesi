```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
# if experiance > 10: performance_score+1 

def adjust_performance(experience):
    if experience >10:
        return 1
    else:
        return 0
```

```python
df["Experience"].apply(adjust_performance)
```

```python
df["Adjusted"] = df["Experience"].apply(adjust_performance)
```

```python
df
```

```python
df["New_Performance_score"] = df["Adjusted"] + df["Performance_Score"]
```

```python
df
```

```python
def adjust_new(row):
    if row["Experience"]>10:
        return row["Performance_Score"]+1
    else:
        return row["Performance_Score"] 
```

```python
df["NewPC"] = df.apply(adjust_new, axis=1) 
```

```python
df
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x : x.replace("_", " "))
```
